```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
randomstringLCSS = open("./analysis/LCSSrandomstring100000.json")
randomstringKMP = open("./analysis/KMPrandomstring100000.json")
iterlist = list(50000 * x for x in range(1,11))
timelistLCSS = json.load(randomstringLCSS)
timelistKMP = json.load(randomstringKMP)
```

```python
plt.plot(iterlist,timelistLCSS, color="red", label="LCSS")
plt.plot(iterlist,timelistKMP, color="blue", label="KMP")
plt.legend()
plt.xlabel("n characters")
plt.ylabel("time (ms)")
```
