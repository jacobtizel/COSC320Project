```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
patternandstring = open("./analysis/KMPpatternanddata500000.json")
iterlistPS = list(50000 * x for x in range(1,51))
timelistPS = json.load(patternandstring)
print(iterlistPS)
print(timelistPS)
patternandstring.close()
```

```python
plt.plot(iterlistPS,timelistPS, color="red", label="increasing pattern length and data length")
plt.legend()
plt.xlabel("n characters")
plt.ylabel("time (ms)")
```

```python
randomstring = open("./analysis/KMPrandomstring5000000.json")
iterlistRS = list(500000 * x for x in range(1,51))
timelistRS = json.load(randomstring)
print(iterlistRS)
print(timelistRS)
randomstring.close()
```

```python
stringlength = open("./analysis/KMPstringlength5000000.json")
iterlistSL = list(500000 * x for x in range(1,51))
timelistSL = json.load(stringlength)
print(iterlistSL)
print(timelistSL)
stringlength.close()
```

```python
plt.plot(iterlistRS,timelistRS, color="red", label="repeatedly adding random letters to data")
plt.plot(iterlistSL,timelistSL, color="blue", label="repeatedly adding pattern to data")
plt.legend()
plt.xlabel("n characters")
plt.ylabel("time (ms)")
```
