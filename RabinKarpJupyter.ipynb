```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
datanumber = open("./analysis/datanumber2000.js", "r")
iterlistDN = list(250 * x for x in range(1, 41))
timelistDN = json.load(datanumber)
print(iterlistDN)
print(timelistDN)
datanumber.close()
```

```python
datasize = open("./analysis/datasize2000.js", "r")
iterlistDS = list(250 * x for x in range(1, 41))
timelistDS = json.load(datasize)
print(iterlistDS)
print(timelistDS)
datasize.close()
```

```python
plt.plot(iterlistDS, timelistDS, color="red", label="increasing data sentence size")
plt.plot(iterlistDN, timelistDN, color="blue", label="increasing data sentence number")
plt.legend()
plt.xlabel("n characters")
plt.ylabel("Time(ms)")
```

```python
filenumber = open("./analysis/filenumber2000.js", "r")
iterlistFN = list(250 * x for x in range(1, 41))
timelistFN = json.load(filenumber)
print(iterlistFN)
print(timelistFN)
filenumber.close()
```

```python
filesize = open("./analysis/filesize2000.js", "r")
iterlistFS = list(250 * x for x in range(1, 41))
timelistFS = json.load(filesize)
print(iterlistFN)
print(timelistFN)
filesize.close()
```

```python
plt.plot(iterlistFS, timelistFS, color="red", label="increasing file sentence size")
plt.plot(iterlistFN, timelistFN, color="blue", label="increasing file sentence number")
plt.legend()
plt.xlabel("n characters")
plt.ylabel("Time(ms)")
```
